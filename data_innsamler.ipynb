```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Last inn CSV-filen
df = pd.read_csv("Antall døde rovdyr per år.csv")

# Sjekk de første radene for å forstå strukturen
print(df.head())

# Plott en graf (tilpass kolonnenavnene etter behov)
df.plot(x="År", y="Antall", kind="line", marker='o') 


plt.xlabel("År")
plt.ylabel("Antall døde rovdyr")
plt.title("Antall døde rovdyr per år")
plt.grid(True)
plt.show()
```
