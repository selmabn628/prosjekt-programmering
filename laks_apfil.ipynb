```python
import requests
import json

url = "https://data.ssb.no/api/v0/dataset/1122.json?lang=no"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()  # Konverterer responsen til et Python-dictionary

    # Skriv ut strukturen for  forst hvordan dataene er lagret
    print(json.dumps(data, indent=2))  # Formaterer JSON-dataene pent

else:
    print(f"Feil ved henting av data: {response.status_code}")


print(data.values())


```

```python
import requests 
import json 
import pandas as pd 
 
# Hente data fra SSB API 
url = "https://data.ssb.no/api/v0/no/table/03024/" 
query = { 
    "query": [], 
    "response": { 
        "format": "json-stat2" 
        } 
} 
 
response = requests.post(url, json=query) 
    
if response.status_code == 200: 
    data = response.json() 
# Ekstrakter variabler fra JSON 
    varegrupper = data["dimension"]["VareGrupper2"]["category"]["label"] 
    tidsperioder = data["dimension"]["Tid"]["category"]["label"] 
    verdier = data["value"] 
 
# Konverter JSON-data til en strukturert liste 
    rows = [] 
    index = 0 
    for tid in tidsperioder.keys(): 
        for vare in varegrupper.keys(): 
            rows.append([tidsperioder[tid], varegrupper[vare], verdier[index]]) 
            index += 1 
 
# Lage en DataFrame 
    df = pd.DataFrame(rows, columns=["Tid", "Varegruppe", "Verdi"]) 
 
# Vise tabellen i Jupyter Notebook eller terminal 
    print(df) 
else: 
    print(f"Feil: {response.status_code}, {response.text}") 
 
```
